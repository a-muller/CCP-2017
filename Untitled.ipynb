```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def filtrer1(filtreA, matB):
    nA = filtreA.shape[0]
    nb_ligneB = matB.shape[0]
    nb_colonneB = matB.shape[1]
    C = matB.copy()
    bordure = (nA - 1) // 2
    for i in range(bordure, nb_ligneB - bordure):
        for j in range(bordure, nb_colonneB - bordure):
            Bij = matB[i-bordure:i+bordure+1, j-bordure:j+bordure+1] 
            C[i,j] = np.sum(np.dot(Bij,filtreA))
    return C
```

```python
def filtrer(filtreA, matB):
    C = matB.copy()
    for i in [0,1,2]:
        C[:,:,i] = filtrer1(filtreA, matB[:,:,i])
    return C
```

```python
def matriceFlouGaussien(taille, sigma):
    mat = np.zeros([taille, taille])
    taille = (taille -1)//2
    for x in range(-taille, taille+1):
        for y in range(-taille, taille+1):
            mat[x+taille, y+taille] = np.exp(- (x**2 + y**2)/ (2*sigma**2))
    mat = mat / np.sum(mat)
    return mat
```

```python
np.round(1002 * matriceFlouGaussien(5,0.9))
```

```python
def FloutageGaussien(tabPix, taille, sigma):
    mat = tabPix.copy()
    filtre = matriceFlouGaussien(taille, sigma)
    flou = filtrer(filtre, mat)
    return flou
```
